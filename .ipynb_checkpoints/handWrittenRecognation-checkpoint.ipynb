```python
#importing packages
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import os
```

## Loading MNIST Dateset
Containing 60,000 Training samples, 10,000 Test samples 

The MNIST database (Modified National Institute of Standards and Technology database) is a large database of handwritten digits that is commonly used for training various image processing systems.

```python
# we can load this dataset using keras
mnist = tf.keras.datasets.mnist
```

## Divide dataset into train and test dataset


```python
path = 'C:/Users/Kp/Documents/HandWrittenRecognation/mnist.npz'
#we can also load dataset manualy in our local machine using numpy by giving the path to it

with np.load(path, allow_pickle=True) as f:
    x_train, y_train = f['x_train'], f['y_train']
    x_test, y_test = f['x_test'], f['y_test']
```

### checking shape of our dataset

```python

print ('x_train shape is : ' +str(x_train.shape))
print ('y_train shape is : ' +str(y_train.shape))
print ('x_test shape is : ' +str(x_test.shape))
print ('y_test shape is : ' +str(y_test.shape))
```

```python
#showing third elemnt in our training set and make it binary
plt.imshow(x_train[2])
plt.show()
plt.imshow(x_train[2], cmap = plt.cm.binary)
```

```python
#printing our image before normalization
print(x_train[2])
```

### Images have 1 channel (0 to 255) and they are not colored
#### Normalizing the data

```python
x_train = tf.keras.utils.normalize(x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_test, axis =1)
plt.imshow(x_train[2], cmap= plt.cm.binary)
```

### After normalizing


```python
print(x_train[2])
#as we can see the amount of data is between 0 to 1 
# what normalizing does is dividing the data by 255 
```

```python
# lets check lable for this specefic index
print(y_train[2])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten
```

## Creating our model


```python
model = Sequential()
#add layers
model.add(Flatten(input_shape=(28, 28)))
 # a layer with 128 units and relu function
model.add(Dense(128, activation = "relu"))
model.add(Dense(128, activation = "relu"))
model.add(Dense(128, activation = "relu"))
model.add(Dense(64, activation = "relu"))
#using softmax to show probability for each digit to be the right answer
model.add(Dense(10, activation ='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

#train model using fit function
model.fit(x_train, y_train, epochs=8)
#saving the model so we can use it again without training again
model.save('handWritten.model')
```

#### Loading model 

```python
model = tf.keras.models.load_model('handWritten.model')
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"loss is :{loss}" )
print(f"accuracy is :{accuracy}")
```

### Testing our model with using drawing digit in paint

```python
image_number = 1;
while os.path.isfile(f"digits/digit{image_number}.png"):
    try:
        #getting first channel because we dont care about colors
        img = cv2.imread(f"digits/digit{image_number}.png")[:,:,0]
        img = np.invert(np.array([img]))
        #make a prediction
        prediction = model.predict(img)
        print(f"This digit is probably a {np.argmax(prediction)}")
        plt.imshow(img[0], cmap = plt.cm.binary)
        plt.show()
        
    except:
        print("Error \n somthing is wrong!")
    finally:
        image_number +=1
```
